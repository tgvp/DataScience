```python
# import graphing libraries
import matplotlib.pyplot as plt

# create subplots
n = 2
fig, axs = plt.subplots(n, n,figsize=(10, 10))

# size of the bins
bins_list = [[1000,500],[100,10]]

# for each row of the subplot
for row in range(0,n):
  #for each column
  for col in range(0,n):
    #fazer o plot no endereo x,y conforme bins[x][y]
    college.Apps.hist(ax=axs[row][col],bins = bins_list[row][col]).set_title('Number of applications')


```
